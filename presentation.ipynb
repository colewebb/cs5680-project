This project is an implementation of the above paper in Python. The paper outlines a four-step process for segmenting a color image and displays favorable results compared to other common segmentation algorithms. The steps are:

 - Transform the color image to YC<sub>b</sub>C<sub>r</sub> color space.
 - Apply an automatic seeding algorithm to generate seeds for the next step.
 - Apply a seeded region growing algorithm to the seeds from the last step.
 - Apply a region-merging algorithm to the regions from the last step.

The steps are implemented as functions in a segment class, in ```segment.py```. Additional tools are in ```tools.py```. In addition to OpenCV, [numpy](https://numpy.org) and [matplotlib](https://matplotlib.org/) are used. This notebook makes use of the [Jupyter](https://jupyter.org/).
